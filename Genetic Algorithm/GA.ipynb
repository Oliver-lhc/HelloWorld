```python
import nbimporter                            #导入.ipynb格式作为模块的工具
import matplotlib.pyplot as plt
from new_pop import gene_encoding
from cal_value import cal_value
from cal_fitness import cal_fitness
from get_best import get_best
from selection import select
from crossover import crossover
from mutation import mutate
```

```python
pop_size = 100          # 种群规模
gene_length = 16        # 基因长度
generation = 200        # 迭代次数
Pc = 0.7                # 交叉概率
Pm = 0.02               # 变异概率

c_max = 10              # x的取值范围
c_min = 0

pop = gene_encoding(pop_size, gene_length)  # 随机初始化种群
X, value = cal_value(pop, c_max, c_min)     # 计算函数值
fitness = cal_fitness(value, c_min)         # 计算适应度
```

```python
gen_x = []
best_y = []

for i in range(generation):
    
    pop = select(pop, pop_size, fitness)    #选择
    pop = crossover(pop, Pc)                #交叉互换
    pop = mutate(pop, Pm)                   #变异
    
    X, value = cal_value(pop, c_max, c_min, gene_length)
    fitness = cal_fitness(value, c_min)
    
    x, y = get_best(fitness, X, value)
    gen_x.append(i)
    best_y.append(y)
    
    print(x,y)
```

```python
plt.plot(gen_x, best_y)
```
