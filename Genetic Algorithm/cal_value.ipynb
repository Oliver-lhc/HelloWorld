```python
import math
import nbimporter
from target_fun import f
```

```python
def decoding(pop):
    """
    解码，计算每条染色体代表的数值
    """
    dec_value = []
    for chrom in pop:
        temp = 0
        for i in range(len(chrom)):
            temp += chrom[i] * math.pow(2, len(chrom)-i-1)
        dec_value.append(temp)
    
    return dec_value
```

```python
def cal_value(pop, c_max, c_min, gene_length):
    """
    计算每条染色体对应的函数值
    """
    dec_value = decoding(pop)
    X = []
    value = []
    for i in range(len(dec_value)):
        temp = 0
        temp = c_min + dec_value[i] * (c_max - c_min)/(math.pow(2, gene_length)-1)
        X.append(temp)
        value.append(f(temp))          #将x代入目标函数
        
    return X, value
```
